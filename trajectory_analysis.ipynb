```python
def draw_hand(kp):
    x = kp[:,0]
    y = 1 - kp[:,1]
    plt.scatter(x,y)
    plt.plot(x[0:5],y[0:5])
    idx = [0] + list(range(5,8+1))
    plt.plot(x[idx],y[idx])
    idx = [0] + list(range(9,12+1))
    plt.plot(x[idx],y[idx])
    idx = [0] + list(range(13,16+1))
    plt.plot(x[idx],y[idx])
    idx = [0] + list(range(17,20+1))
    plt.plot(x[idx],y[idx])
    idx = [5,9,13,17]
    plt.plot(x[idx],y[idx])
    plt.xlim(0,1)
    plt.ylim(0,1)

plt.figure()
draw_hand(hand_1_traj)
plt.show()
```

```python
import cv2

label_path = "../datasets/hand-keypoints/train/labels/IMG_00000001.txt"
img_path = "../datasets/hand-keypoints/train/IMG_00000001.jpg"

with open(label_path, 'r') as f:
    label = f.read().split()

label = np.array(label[5:], np.float32)
label = label.reshape((21,3))
print(label)

plt.figure()
draw_hand(label[:,:-1])
plt.show()
```
