```python
from ultralytics import YOLO
import torch
import torchvision

torch.cuda.is_available()
```

```python
model = YOLO("yolo11n-pose.pt")

results = model.train(
    data="hand-keypoints.yaml", 
    epochs=20, 
    imgsz=640, 
    device='cuda',
    erasing=0.4)
```

```python
print(results)
```
